```python
import discord

intents = discord.Intents.default()
intents.messages = True  # Ensure messages intent is enabled
intents.guilds = True  # Ensure guilds intent is enabled

TOKEN = 'MTIxMzc4ODQ5ODcxNzMxNTExMw.GW9vK6.5CSqpDqLDZxcH6hTmWoZUserPTP2yoNyvH-Y-Y'  # Replace this with your actual bot token
THREAD_ID = 1205565888183668736  # Make sure this is the correct thread ID
messages = []

class MyBot(discord.Client):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs, intents=intents)
        
    async def on_ready(self):
        print(f'Logged in as {self.user}')
        thread = await self.fetch_channel(THREAD_ID)
        if isinstance(thread, discord.Thread):
            print(f"Found the thread: {thread.name}, fetching messages...")
            async for message in thread.history(limit=None):
                print(f'{message.author}: {message.content} (at {message.created_at})')
                messages.append([message.author.name, message.content, message.created_at])
        else:
            print("Provided ID does not belong to a thread.")

bot = MyBot()
bot.run(TOKEN)
```

```python
df = pd.DataFrame(messages, columns=["name", "message", "timestamp"]).query("message == '💩'")
df["date"] = pd.to_datetime(df["timestamp"].dt.date)
df["hour"] = df["timestamp"].dt.hour
df.head()
```

```python
alt.Chart(df).mark_circle(size=200).encode(
    x=alt.X("date:T", title= ""),
    y=alt.Y("hour:N", title="Hora"),
    color=alt.Color("name", title="")
).properties(width=1000, height=500)
```

```python
alt.Chart(df).mark_bar().encode(
    y=alt.Y('week(date):O', title = "Semana" ),
    x=alt.X('count():Q', title = "Cacas"),
    color=alt.Color("name", title="")
).properties(width=700)
```

```python
df_count_per_day = df.groupby(["name", "date"]).count()
df_count_per_day.groupby("name").mean().reset_index()[["name", "message"]]
```
